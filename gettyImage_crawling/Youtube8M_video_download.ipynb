## Youtube8M feature 확인 가능 & class, video id 확인

```python
# Video-level information (extracted from the frame-level files)

frame_lvl_record = f'{CFG.tf_record_dir}/{CFG.tf_record}.tfrecord'
vid_ids = []
labels = []

for example in tf.compat.v1.python_io.tf_record_iterator(frame_lvl_record):
    tf_example = tf.train.Example.FromString(example)
    vid_ids.append(tf_example.features.feature['id']
                   .bytes_list.value[0].decode(encoding='UTF-8'))
    labels.append(tf_example.features.feature['labels'].int64_list.value)
    
print('Number of videos in this tfrecord: ',len(vid_ids))
# print ('Number of labels in this tfrecord: ', len (labels))
```

```python
def readTfRecordSamples(tfrecords_filename):
    video_ids = []
    record_iterator =tf.compat.v1.python_io.tf_record_iterator(path=tfrecords_filename)

    for string_record in record_iterator:

        example = tf.train.SequenceExample()
        example.ParseFromString(string_record)
#         print(f"Example :\n{example}")

        img_string = (example.context.feature)
#         print ("Features are : \n")
#         print(img_string)
        classID = (example.context.feature['labels'].int64_list.value[0])
        videoID = (example.context.feature['id'].bytes_list.value[0])
        
        print (f"class: {classID}, video: {videoID}")
        video_ids.append(str(videoID).replace("b","").replace("'",""))
    
        rgbArray = (example.feature_lists.feature_list['rgb'].feature)
#         print(rgbArray)
#         raw_input(LineSeperator)
    return video_ids

# tf_record에 담겨있는 video들의 id
video_ids = readTfRecordSamples(f'{CFG.tf_record_dir}/{CFG.tf_record}.tfrecord')
```

## 원본 동영상 보기

```python
def original_video(hidden):
    url = requests.get(f"http://data.yt8m.org/2/j/i/{hidden[:2]}/{hidden}.js")
    text = url.text
    return text.split('"')[-2]
```

## 영상 다운로드

```python
from pytube import YouTube
def download_video(video_id):
    download_path = CFG.download_path
    video_url = f"https://www.youtube.com/watch?v={original_video(video_id)}"
    yt = YouTube(video_url)
    stream = yt.streams.get_highest_resolution()
    stream.download(download_path)

    print("제목 : ", yt.title)
    print("길이 : ", yt.length)
    print("게시자 : ", yt.author)
    print("게시날짜 : ", yt.publish_date)
    print("조회수 : ", yt.views)
    print("키워드 : ", yt.keywords)
    print("설명 : ", yt.description)
    print("썸네일 : ", yt.thumbnail_url)
```
